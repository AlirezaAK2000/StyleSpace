```python
from tqdm import tqdm
```

```python
with open('npy/ffhq/S' , 'rb') as f:
    S = pickle.load(f)
```

```python
S = np.concatenate(S[1],axis=1)
S.shape
```

```python
from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.metrics.pairwise import pairwise_distances_argmin , pairwise_distances
```

```python
# Load data in X 
batch_size = 1000
  
# perform the mini batch K-means
mbk = MiniBatchKMeans(init ='k-means++', n_clusters = 5,
                      batch_size = batch_size,
                      max_no_improvement = 10, verbose = 1)
  
mbk.fit(S)
centers = mbk.cluster_centers_
labels = pairwise_distances_argmin(S, centers)
```

```python
from collections import defaultdict
from copy import deepcopy
import pandas as pd
```

```python
ffhq = pd.read_csv('npy/ffhq/attribute').keys()
ffhq = defaultdict(list,{ k:0 for k in ffhq })
```

```python
for i in unique:
    imgs = np.where(labels == i)[0]
    f = deepcopy(ffhq)
    for img in imgs:
        for k in img_cls_ffhq[str(img)]:
            f[k] += 1
    print(min(f.values()))
    plt.figure(figsize =(100, 7))
    plt.bar(f.keys() , f.values())
    plt.title(i + 1)
    plt.show()
    
```
